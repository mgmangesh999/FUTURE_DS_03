```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../Data/cleaned_marketing_funnel.csv")
df.head()
```

### Analyze Funnel Stages

```python
## Count Funnel Events
funnel = df['event_type'].value_counts()
print(funnel)
```

### Calculate Conversion Rates

```python
# Extract Funnel Counts
views = len(df[df['event_type'] == 'view'])
cart = len(df[df['event_type'] == 'cart'])
purchase = len(df[df['event_type'] == 'purchase'])
```

```python
# Visitor --> Lead Conversion Rate
visitor_to_lead = (cart/views) * 100
print(visitor_to_lead)
```

```python
# Lead --> Customer Conversion Rate
lead_to_customer = (purchase/cart) * 100
print(lead_to_customer)
```

```python
# Overall Funnel conversion Rate
overall_conversion = (purchase/views) * 100
print(overall_conversion)
```

### Analyze Drop-Off rates

```python
view_drop = 100 - visitor_to_lead
cart_drop = 100 - lead_to_customer
```

```python
print("Visitor drop-off : ", view_drop)
print("Lead drop-off : ", cart_drop)
```

### Analyze Top categories

```python
# Most viewed categories
views_by_category = (df[df['event_type'] == 'view']['category_code'].value_counts().head())
print(views_by_category)
```

```python
# Most Purchase Categories
purchase_by_category = (df[df['event_type'] == 'purchase']['category_code'].value_counts().head())
print(purchase_by_category)
```

### Analyze Brand Performance

```python
brand_sales = (df[df['event_type'] == 'purchase'].groupby('brand')['price'].sum().sort_values(ascending=False).head(10))
print(brand_sales)
```

### Daily Funnel Activity

```python
activity = (df.groupby(['date','event_type']).size().unstack(fill_value=0))
print(activity)
```
